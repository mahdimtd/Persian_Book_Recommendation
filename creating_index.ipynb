```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
```

```python
import pandas as pd

books = pd.read_csv('/content/cleaned_data.csv')
```

```python
raw_document = TextLoader('/content/tagged_description.txt').load()
```

```python
text_splitter = CharacterTextSplitter(chunk_size=0,chunk_overlap=0,separator='\n')
```

```python
documents = text_splitter.split_documents(raw_document)
```

```python
documents[1000]
```

```python
embed_model = HuggingFaceEmbeddings(model_name='PartAI/Tooka-SBERT-V2-Large')
```

```python
books_db = Chroma.from_documents(
    documents,
    embedding = embed_model
)
```

```python

query = "    "
docs = books_db.similarity_search(query, k = 10)
docs
```

```python
import re
import pandas as pd

def search_books(query, books_db, books, k=10):
    docs = books_db.similarity_search(query, k=k)

    book_rows = []
    for doc in docs:
        text = doc.page_content.strip()
        match = re.match(r"(\d+)", text)
        if match:
            book_id = int(match.group(1))
            row = books[books["book_id"] == book_id]
            if not row.empty:
                book_rows.append(row)

    if book_rows:
        result_df = pd.concat(book_rows, ignore_index=True)
    else:
        result_df = pd.DataFrame()
    return result_df

```

```python
query = '      '
result_df = search_books(query, books_db, books, k=3)
display(result_df)
```
